```python
import pandas as pd

# Load the dataset
df = pd.read_csv('union.csv')

# Calculate the Pearson correlation coefficient
correlation = df['CPM'].corr(df['Altitude'])

print(f"The correlation coefficient between CPM and Altitude is: {correlation}")
```

```python
import pandas as pd
import scipy.stats as stats

df = pd.read_csv('union.csv')

# 1. Normality Test
print("Shapiro-Wilk (CPM) p-value:", stats.shapiro(df['CPM']).pvalue)

# 2. Spearman Rank Correlation
corr, p = stats.spearmanr(df['CPM'], df['Altitude'])
print(f"Spearman Correlation: {corr:.3f}, p-value: {p:.5f}")

# 3. Linear Regression
slope, intercept, r_val, p_val, std_err = stats.linregress(df['Altitude'], df['CPM'])
print(f"Regression Line: CPM = {slope:.4f}*Alt + {intercept:.2f}")
print(f"R-squared: {r_val**2:.3f}")
```
